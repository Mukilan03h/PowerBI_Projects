```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Load the CSV file
df = pd.read_excel('D:\\CubeAI Internship\\Task-5 Road Accident EDA\\Road Accident Data.xlsx')
```

```python
# First few rows of the dataframe
print(df.head())
```

```python
# Summary of Data Frame
print(df.info())
```

```python
# Basic statistics for numerical columns
print(df.describe(include='all'))
```

```python
# Statistics for numerical columns
print(df.describe())

# Statistics for categorical columns
print(df.describe(include=[object]))
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Percentage of missing values in each column
missing_values = df.isnull().mean() * 100
print(missing_values)

# Visualizing missing values
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=True, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python

# Distribution of accidents by severity
plt.figure(figsize=(10, 6))
sns.countplot(x='Accident_Severity', data=df, palette='viridis')
plt.title('Distribution of Accidents by Severity')
plt.show()

```

```python
# Distribution of accidents by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_Week', data=df, palette='viridis', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Distribution of Accidents by Day of the Week')
plt.show()



```

```python
# Distribution of accidents by light conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Light_Conditions', data=df, palette='viridis')
plt.title('Distribution of Accidents by Light Conditions')
plt.xticks(rotation=45)
plt.show()
```

```python

# Distribution of accidents by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Conditions', data=df, palette='viridis')
plt.title('Distribution of Accidents by Weather Conditions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of accidents by road surface conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Surface_Conditions', data=df, palette='viridis')
plt.title('Distribution of Accidents by Road Surface Conditions')
plt.xticks(rotation=45)
plt.show()
```

```python

# Distribution of accidents by road type
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Type', data=df, palette='viridis')
plt.title('Distribution of Accidents by Road Type')
plt.xticks(rotation=45)
plt.show()
```

```python

# Distribution of accidents by speed limit
plt.figure(figsize=(10, 6))
sns.countplot(x='Speed_limit', data=df, palette='viridis')
plt.title('Distribution of Accidents by Speed Limit')
plt.xticks(rotation=45)
plt.show()

```

```python
# Distribution of accidents by urban or rural area
plt.figure(figsize=(10, 6))
sns.countplot(x='Urban_or_Rural_Area', data=df, palette='viridis')
plt.title('Distribution of Accidents by Urban or Rural Area')
plt.show()

```

```python
# Distribution of accidents by vehicle type
plt.figure(figsize=(12, 6))
sns.countplot(x='Vehicle_Type', data=df, palette='viridis')
plt.title('Distribution of Accidents by Vehicle Type')
plt.xticks(rotation=90)  # Rotate x-axis labels for better alignment
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
top_cities = df.groupby('Local_Authority_(District)')['Number_of_Casualties'].sum().reset_index()

top_cities = top_cities.sort_values(by='Number_of_Casualties', ascending=False)

top_10_cities = top_cities.head(10)

# Visualize the top 10 cities by number of casualties
plt.figure(figsize=(12, 6))
sns.barplot(x='Local_Authority_(District)', y='Number_of_Casualties', data=top_10_cities, palette='viridis')
plt.title('Top 10 District by Number of Casualties')
plt.xlabel('Local Authority District')
plt.ylabel('Number of Casualties')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```
