```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Install & import required Python libraries
# Already done above

# 2. Download the dataset and convert it into a Pandas DataFrame
# Assuming the dataset is already downloaded and placed in the specified path
swiggy_df = pd.read_csv('D:\\CubeAI Internship\\Task-9 Swiggy Data Analysis\\swiggy.csv')

# 3. Data Preparation & Cleaning The Dataset
# Display a concise summary of the DataFrame
print(swiggy_df.info())

# Create a copy of the DataFrame to keep the original data safe
swiggy1_df = swiggy_df.copy()

# Check for null values
print(swiggy1_df.isnull().sum())

# Drop rows with any null values
swiggy1_df = swiggy1_df.dropna()

# Convert 'Avg ratings' to float, 'Total ratings' and 'Price' to int
swiggy1_df['Avg ratings'] = swiggy1_df['Avg ratings'].astype(float)
swiggy1_df['Total ratings'] = swiggy1_df['Total ratings'].astype(int)
swiggy1_df['Price'] = swiggy1_df['Price'].astype(int)

# 4. Perform exploratory analysis & visualization

# a. Average Price of Food Items
average_price = swiggy1_df['Price'].mean()
print("Average Price of Food Items:", average_price)

plt.figure(figsize=(6,4))
plt.bar(['Average Price'], [average_price], color='skyblue')
plt.ylabel('Price')
plt.title('Average Price of Food Items')
plt.show()

# b. Area with the Highest Number of Restaurants
area_counts = swiggy1_df['Area'].value_counts()
print("Area with the highest number of restaurants:", area_counts.idxmax())

plt.figure(figsize=(10,6))
area_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Restaurants per Area')
plt.xlabel('Area')
plt.ylabel('Number of Restaurants')
plt.show()

# c. Distribution of Restaurants across Different Cities
city_counts = swiggy1_df['City'].value_counts()
print("Distribution of restaurants across different cities:\n", city_counts)

plt.figure(figsize=(8,8))
city_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Restaurants across Cities')
plt.ylabel('')
plt.show()

# d. Average Rating of Restaurants
average_rating = swiggy1_df['Avg ratings'].mean()
print("Average Rating of Restaurants:", average_rating)

plt.figure(figsize=(6,4))
plt.bar(['Average Rating'], [average_rating], color='lightgreen')
plt.ylabel('Average Rating')
plt.title('Average Rating of Restaurants')
plt.show()

# e. Correlation between Price and Average Ratings
correlation = swiggy1_df['Price'].corr(swiggy1_df['Avg ratings'])
print("Correlation between price and average ratings:", correlation)

plt.figure(figsize=(8,6))
sns.scatterplot(x='Price', y='Avg ratings', data=swiggy1_df, color='orange')
plt.title('Price vs Average Ratings')
plt.xlabel('Price')
plt.ylabel('Average Ratings')
plt.show()

# f. Most Popular Food Types
a = []
for i in swiggy1_df['Food type'].values:
    try:
        for j in i.split(','):
            a.append(j.strip())  # Strip whitespace from each food type
    except:
        a.append(i) 

popular_food_df = pd.DataFrame({"all_food_types": a}).value_counts().head(10)
print("Most popular food type:", popular_food_df.index[0][0])

plt.figure(figsize=(10,6))
popular_food_df.plot(kind='bar', color='salmon')
plt.title('Top 10 Most Popular Food Types')
plt.xlabel('Food Type')
plt.ylabel('Frequency')
plt.show()

# g. Average Delivery Time across Different Areas
average_delivery_time_by_area = swiggy1_df.groupby('Area')['Delivery time'].mean()
print("Average delivery time by area:\n", average_delivery_time_by_area)

plt.figure(figsize=(10,6))
average_delivery_time_by_area.plot(kind='bar', color='purple')
plt.title('Average Delivery Time across Different Areas')
plt.xlabel('Area')
plt.ylabel('Average Delivery Time (minutes)')
plt.show()

# h. Relationship between Total Ratings and Average Ratings
correlation_total_avg_ratings = swiggy1_df['Total ratings'].corr(swiggy1_df['Avg ratings'])
print("Correlation between total ratings and average ratings:", correlation_total_avg_ratings)

plt.figure(figsize=(8,6))
sns.scatterplot(x='Total ratings', y='Avg ratings', data=swiggy1_df, color='blue')
plt.title('Total Ratings vs Average Ratings')
plt.xlabel('Total Ratings')
plt.ylabel('Average Ratings')
plt.show()

# i. Most Common Food Type per Area
most_common_food_type_per_area = swiggy1_df.groupby('Area')['Food type'].apply(lambda x: pd.Series(','.join(x).split(',')).value_counts().idxmax())

# Convert the series to a DataFrame to plot
most_common_food_type_per_area_df = most_common_food_type_per_area.reset_index()
most_common_food_type_per_area_df.columns = ['Area', 'Most Common Food Type']

# We can't directly plot categorical data. Let's count occurrences of the most common food types across areas.
food_type_counts = most_common_food_type_per_area_df['Most Common Food Type'].value_counts()

print("Most common food type per area:\n", most_common_food_type_per_area)

plt.figure(figsize=(10,6))
food_type_counts.plot(kind='bar', color='lightcoral')
plt.title('Most Common Food Type per Area')
plt.xlabel('Food Type')
plt.ylabel('Frequency')
plt.show()

# j. Average Rating vs Delivery Time
average_rating_by_delivery_time = swiggy1_df.groupby('Delivery time')['Avg ratings'].mean()
print("Average rating by delivery time:\n", average_rating_by_delivery_time)

plt.figure(figsize=(10,6))
average_rating_by_delivery_time.plot(kind='line', marker='o', color='teal')
plt.title('Average Rating vs Delivery Time')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Average Rating')
plt.show()

# 5. Asking And Answering Questions from Dataset

# a. Top 10 Cities as per Number of Restaurants Listed
top_cities = swiggy1_df['City'].value_counts().head(10)
print("Top 10 cities as per number of restaurants listed:\n", top_cities)

plt.figure(figsize=(10,6))
top_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()

# b. Restaurants with Maximum Number of Branches
restaurant_branches = swiggy1_df['Restaurant'].value_counts().head(10)
print("Restaurants with the maximum number of branches:\n", restaurant_branches)

plt.figure(figsize=(10,6))
restaurant_branches.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Restaurants by Number of Branches')
plt.xlabel('Restaurant')
plt.ylabel('Number of Branches')
plt.show()

# c. Cities with Most Expensive Food
average_cost_per_city = swiggy1_df.groupby('City')['Price'].mean().sort_values(ascending=False).head(10)
print("Cities with the most expensive food:\n", average_cost_per_city)

plt.figure(figsize=(10,6))
average_cost_per_city.plot(kind='bar', color='salmon')
plt.title('Top 10 Cities with Most Expensive Food')
plt.xlabel('City')
plt.ylabel('Average Cost')
plt.show()

# Top 5 Most Popular Restaurant Chains in Terms of Ratings
rating_count_100_df = swiggy1_df[swiggy1_df['Total ratings'] >= 100]
top_5_popular_chains_df = rating_count_100_df.groupby('Restaurant')['Avg ratings'].mean().sort_values(ascending=False).head(5)
print("Top 5 most popular restaurant chains in terms of ratings:\n", top_5_popular_chains_df)

plt.figure(figsize=(10,6))
top_5_popular_chains_df.plot(kind='bar', color='purple')
plt.title('Top 5 Most Popular Restaurant Chains')
plt.xlabel('Restaurant')
plt.ylabel('Average Rating')
plt.show()

# Most Popular Cuisines Served
most_popular_cuisines = pd.DataFrame({"all_food_types": a}).value_counts().head(10)
print("Most popular cuisines:\n", most_popular_cuisines)

plt.figure(figsize=(10,6))
most_popular_cuisines.plot(kind='bar', color='lightblue')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Frequency')
plt.show()
```
