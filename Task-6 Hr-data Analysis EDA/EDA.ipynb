```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('D:\\CubeAI Internship\\Task-6 Hr-data Analysis EDA\\hr-data.xlsx')
```

```python
# Display information about the dataframe
print("Information about the dataframe:")
print(df.info())
```

```python
# Display summary statistics for numeric columns
print("\nSummary Statistics for numeric columns:")
print(df.describe())
```

```python

# Display summary statistics for categorical columns
print("\nSummary Statistics for categorical columns:")
print(df.describe(include=['object']))
```

```python
# Frequency of Gender
print("\nFrequency of Gender:")
print(df['Gender'].value_counts())

# Frequency of Education Qualification
print("\nFrequency of Education Qualification:")
print(df['Education Qualification'].value_counts())
```

```python
# Plotting age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

df['Date of Join'] = pd.to_datetime(df['Date of Join'])
df['Salary'] = df['Salary'].replace('[\\$,]', '', regex=True).astype(float)
df['Age'] = df['Age'].astype(float)
```

```python

df['Tenure'] = (pd.to_datetime('today') - df['Date of Join']).dt.days / 365

gender_distribution = df['Gender'].value_counts()
job_title_distribution = df['Job Title'].value_counts()
education_distribution = df['Education Qualification'].value_counts()
salary_by_job_title = df.groupby('Job Title')['Salary'].mean()
salary_by_gender = df.groupby('Gender')['Salary'].mean()
hiring_trend = df['Date of Join'].dt.year.value_counts().sort_index()
salary_by_education = df.groupby('Education Qualification')['Salary'].mean()

```

```python

# Visualization
plt.figure(figsize=(30, 20))

# 1. Gender distribution
plt.subplot(3, 3, 1)
gender_distribution.plot(kind='bar', color=['blue', 'orange'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
# 2. Job title distribution
plt.subplot(1, 1, 1)
job_title_distribution.plot(kind='bar', color='green')
plt.title('Job Title Distribution')
plt.xlabel('Job Title')
plt.ylabel('Count')
```

```python
# 3. Education qualifications distribution
plt.subplot(1, 1, 1)
education_distribution.plot(kind='bar', color='purple')
plt.title('Education Qualifications Distribution')
plt.xlabel('Education Qualification')
plt.ylabel('Count')
```

```python
# 4. Salary by job title
plt.subplot(1, 1, 1)
salary_by_job_title.plot(kind='bar', color='red')
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
```

```python
# 5. Salary by gender
plt.subplot(1, 1, 1)
salary_by_gender.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
```

```python
# 6. Hiring trend over time
plt.subplot(1, 1, 1)
hiring_trend.plot(kind='line', marker='o', color='brown')
plt.title('Hiring Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Hires')
```

```python
# 7. Average Salary by Education Qualification
plt.subplot(1, 1, 1)
salary_by_education.plot(kind='bar', color='cyan')
plt.title('Average Salary by Education Qualification')
plt.xlabel('Education Qualification')
plt.ylabel('Average Salary')
```
